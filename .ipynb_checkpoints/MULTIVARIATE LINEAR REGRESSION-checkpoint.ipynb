### importting modules

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from statsmodels.formula.api import ols, glm
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
from math import sqrt
import numpy as np
from sklearn.metrics import mean_squared_error
sns.set()
```

```python
data = pd.read_csv('salary.csv')
data.info()
d_ata = pd.read_csv('Salary metadata.csv', sep=";")
```

```python
a = data[data.yearsworked == 21]
meaan = a[a.position == 3].salary.mean()
data.salary.fillna(meaan, inplace=True)
```

```python
data.info()
```

```python
train_data, test_data  = train_test_split(data, random_state=23) 
```

```python
# train_data get dummies
train_data['categorical'] = pd.Categorical(train_data['Field'])
train_dum = pd.get_dummies(train_data['categorical'], prefix = 'category')
train_data['position'] = pd.Categorical(train_data['position'])
train_dumm = pd.get_dummies(train_data['position'], prefix = 'position') 
train_data = pd.concat([train_data, train_dum, train_dumm], axis=1)

# test_data get dummies
test_data['categorical'] = pd.Categorical(test_data['Field'])
test_dum = pd.get_dummies(test_data['categorical'], prefix = 'category')
test_data['position'] = pd.Categorical(test_data['position'])
test_dumm = pd.get_dummies(test_data['position'], prefix = 'position') 
test_data = pd.concat([test_data, test_dum, test_dumm], axis=1)

train_data.info()
```

```python
mult_model = smf.ols(formula='salary ~ exprior + yearsabs + degree + otherqual + yearsworked + yearsrank + male + market + category_1 + category_2 + category_3 + category_4 + position_1 + position_2 + position_3', data=train_data).fit()
mult_model.summary()
```

```python
prediction = mult_model.predict(train_data)
pred_test = mult_model.predict(test_data)
```

```python
residual = mult_model.resid
fitted = mult_model.fittedvalues
```

```python
sns.residplot(prediction,residual)
plt.title('Residual vs Predictions')
plt.xlabel('Residual')
plt.ylabel('Predictions')
plt.show()
```

```python
rmse_train = sqrt(mean_squared_error(train_data.salary, prediction))
rmse_test = sqrt(mean_squared_error(test_data.salary, pred_test))
print('RMSE Train', rmse_train)
print("RMSE Test", rmse_test)
print("difference between training and test is", rmse_train - rmse_test)
```
